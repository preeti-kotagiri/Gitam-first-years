### Functions: Built-in and user-defined
### Built-in-->examples:print(),input(),range()
### user-defined -->
#### syntax: def name_of_the_function(variable names):
####                   .......................
####                   return statement (optional)


```python
# Addition of two numbers function
def Addition(x,y):#definition
    sum = x + y
    return sum 
print(Addition(10,20))#function calling
```

```python
# finding area and perimeter using functions
def Area(r):
    return 3.14*r*r
def Perimeter(r):
    return 4*3.14*r
print(Area(7))
print(Perimeter(7))
    
```

```python
# square and cube of a number
# x-->5  square-->x*x cube -->x*square(x)
def square(x):
    print(x*x)
def cube(x):
    print(x*x*x)
square(5)
cube(5)
```
